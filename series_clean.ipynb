```python
series.drop(['link','popularity'],axis=1,inplace=True)  #dropping the columns which are not required
```

```python
series["votes"]=series["votes"].str.replace(",","").astype(int) #removing the commas and converting the datatype to float
```

```python
series["duration"].value_counts #plotting the histogram for duration
```

```python
series.drop(["duration"],axis=1,inplace=True)  #dropping the columns which are not required
```

```python
countrylist=[]
titles=[]
country=series["country"].dropna()
for i in range(len(country)):
    x=country.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            countrylist.append(j)
            titles.append(series["title"].iloc[i])
    else:
        countrylist.append(x[0])  
        titles.append(series["title"].iloc[i])
country_df=pd.DataFrame({"title":titles,"country":countrylist})
country_df["country"].head()        

```

```python
directorlsit=[]
titles=[]

director=series["creators"].dropna()
for i in range(len(director)):
    x=director.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            directorlsit.append(j)
            titles.append(series["title"].iloc[i])
            
    else:
        directorlsit.append(x[0])
        titles.append(series["title"].iloc[i])
director_df=pd.DataFrame({"title":titles,"director":directorlsit})
director_df
```

```python
starslist=[]
titles=[]
stars=series["starts"].dropna()
for i in range(len(stars)):
    x=stars.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            starslist.append(j)
            titles.append(series["title"].iloc[i])
    else:
        starslist.append(x[0])
        titles.append(series["title"].iloc[i])
stars_df=pd.DataFrame({"title":titles,"stars":starslist})
stars_df
```

```python
production_companylist=[]
titles=[]
production_company=series["production_company"].dropna()
for i in range(len(production_company)):
    x=production_company.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            production_companylist.append(j)
            titles.append(series["title"].iloc[i])
    else:
        production_companylist.append(x[0])
        titles.append(series["title"].iloc[i])
production_company=pd.DataFrame({"title":titles,"production_company":production_companylist})
production_company
```

```python
language_list=[]
titles=[]
language=series["language"].dropna()
for i in range(len(language)):
    x=language.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            language_list.append(j)
            titles.append(series["title"].iloc[i])
    else:
        language_list.append(x[0])
        titles.append(series["title"].iloc[i])
language=pd.DataFrame({"title":titles,"language":language_list}).head(10)
language
```

```python
genrelsit=[]
titles=[]
genre=series["genre"].dropna()
for i in range(len(genre)):
    x=genre.str.split(",").iloc[i]
    if len(x)>1:
        for j in x:
            genrelsit.append(j)
            titles.append(series["title"].iloc[i])    
    else:
        genrelsit.append(x[0])
        titles.append(series["title"].iloc[i])
genre_df=pd.DataFrame({"title":titles,"genre":genrelsit})
genre_df.head(10)   
```

```python
years_start=[]
years_end=[]
year=series["year"]

for i in range(len(year)):
    if "–" in year.iloc[i][-1]:
        years_start.append(year.iloc[i][0:4])
        years_end.append("unfinished")

    elif "–" in year.iloc[i]:
        x=year.iloc[i].replace("–",",").split(",")
        years_start.append(x[0])
        years_end.append(x[1])
    else:
        years_start.append(year.iloc[i])
        years_end.append("unfinished")
   
series["end_year"]=[ i for i in years_end if i!=""]
series["start_year"]=years_start

series.drop(["year"],axis=1)

```
